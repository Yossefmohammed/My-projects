```python
#remove the outliers:
import numpy as np
filter_data=quality.iloc[:,[0,1,2,3,4,5]]
filter_data.head()
   
```

-In this cell we make a new data frame and take the columns have outliers to remove all  outliers in each column or reduce it we can show that in the coming cells


-We use the inter_quartile range (IQR) to remove the outliers:

```python
for column in filter_data.columns:
    Q1 = filter_data[column].quantile(0.25)
    Q3 = filter_data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    quality[column] = filter_data[(filter_data[column] >= lower_bound) & (filter_data[column] <= upper_bound)][column]
```
