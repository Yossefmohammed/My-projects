```python
import numpy as np
import tensorflow as tf
import wikipediaapi
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense, Dropout
from tensorflow.keras.utils import to_categorical
```

```python
# Fetch Wikipedia article
def fetch_wikipedia_text(topic):
      wiki_wiki = wikipediaapi.Wikipedia(
      language='en',
      extract_format=wikipediaapi.ExtractFormat.WIKI,
      user_agent='my-application/1.0')
      page = wiki_wiki.page(topic)
      return page.text

# Preprocess text
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove special characters
    text = ''.join(c for c in text if c.isalnum() or c in [' ', '\n'])
    return text
```

```python
topic = "Artificial intelligence"
wikipedia_text = fetch_wikipedia_text(topic)
preprocessed_text = preprocess_text(wikipedia_text)
print(preprocessed_text)
```

```python
# Create character-to-integer mapping
chars = sorted(list(set(preprocessed_text)))
char_to_int = {c: i for i, c in enumerate(chars)}
int_to_char = {i: c for i, c in enumerate(chars)}

# Convert characters to numerical representations
encoded_text = [char_to_int[c] for c in preprocessed_text]

# Define sequence length
seq_length = 100
# Generate sequences
sequences = []
for i in range(0, len(encoded_text) - seq_length, 1):
    sequences.append(encoded_text[i:i + seq_length + 1])

# Convert sequences to numpy arrays
X = np.array([seq[:-1] for seq in sequences])
y = np.array([seq[-1] for seq in sequences])
print(X )
print("------------------------------------------------------------------------------------------")
print(y)
```

```python
# Build the model
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(len(chars), 50, input_length=seq_length),
    tf.keras.layers.SimpleRNN(256, return_sequences=True),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.SimpleRNN(256),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(len(chars), activation='softmax')
])

# Compile the model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam')

# Train the model
model.fit(X, y, batch_size=64, epochs=50)
```

```python
def generate_text_base_char(model, seed_text, num_chars=100, seq_length=100):
    # Preprocess the seed text
    seed_text = preprocess_text(seed_text)
    generated_text = seed_text

    for _ in range(num_chars):
        # Convert seed text to numerical representation
        encoded_seed = [char_to_int.get(char, 0) for char in seed_text]  # Use `get` with a default value if char not found

        # Padding if necessary
        if len(encoded_seed) < seq_length:
            padding = [0] * (seq_length - len(encoded_seed))  # Assuming '0' is used for padding
            encoded_seed = padding + encoded_seed

        encoded_seed = np.array([encoded_seed[-seq_length:]])  # Ensure input is exactly `seq_length` long

        # Predict the next character
        predicted_probs = model.predict(encoded_seed)[0]
        predicted_index = np.random.choice(len(predicted_probs), p=predicted_probs)
        predicted_char = int_to_char[predicted_index]

        # Add predicted character to the generated text
        generated_text += predicted_char

        # Update seed text for next prediction
        seed_text += predicted_char
        seed_text = seed_text[1:]  # Sliding window

    return generated_text

seed_text = "Neural"
generated_text = generate_text_base_char(model, seed_text)
print(generated_text)
```

```python
text = preprocessed_text
# Prepare the tokenizer and encode the words
tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])
encoded = tokenizer.texts_to_sequences([text])[0]
vocab_size = len(tokenizer.word_index) + 1

# Generate sequences
seq_length = 100
sequences = []
for i in range(seq_length, len(encoded)):
    seq = encoded[i-seq_length:i+1]
    sequences.append(seq)

sequences = np.array(sequences)
X, y = sequences[:,:-1], sequences[:,-1]
y = to_categorical(y, num_classes=vocab_size)




```

```python
# Define the RNN model
model = Sequential([
    Embedding(vocab_size, 50, input_length=seq_length),
    SimpleRNN(256, return_sequences=True),
    Dropout(0.5),
    SimpleRNN(256),
    Dropout(0.5),
    Dense(vocab_size, activation='softmax')
])

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(X, y, batch_size=64, epochs=150)
```

```python
# Generate text function
def generate_text_base_word(model, tokenizer, seed_text, num_words):
    result = []
    in_text = seed_text
    for _ in range(num_words):
        encoded = tokenizer.texts_to_sequences([in_text])[0]
        encoded = pad_sequences([encoded], maxlen=seq_length, truncating='pre')
        yhat = model.predict(encoded, verbose=0)
        yhat = np.argmax(yhat, axis=1)
        out_word = ''
        for word, index in tokenizer.word_index.items():
            if index == yhat:
                out_word = word
                break
        in_text += ' ' + out_word
        result.append(out_word)
    return ' '.join(result)

# Example usage
seed_text = "Neural"
generated_text = generate_text_base_word(model, tokenizer, seed_text, 50)
print(seed_text,generated_text)
```
